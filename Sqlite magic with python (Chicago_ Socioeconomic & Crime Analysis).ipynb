```python
#List 5 community areas with highest % of households below poverty 
%sql sqlite:///census_data.db
%sql select COMMUNITY_AREA_NAME, PERCENT_HOUSEHOLDS_BELOW_POVERTY \
    from census_data \
    order by PERCENT_HOUSEHOLDS_BELOW_POVERTY desc \
    limit 5
```

```python
#Which community area is most crime prone? Display the coumminty area number only.
%sql sqlite:///crime_data.db
%sql select COMMUNITY_AREA_NUMBER, COUNT(*) AS crime_rate \
    from crime_data \
    where COMMUNITY_AREA_NUMBER != 'None' \
    group by COMMUNITY_AREA_NUMBER \
    order by crime_rate desc \
    limit 1
```

```python
#Use a sub-query to find the name of the community area with highest hardship index
%sql sqlite:///census_data.db
%sql select COMMUNITY_AREA_NAME from census_data where HARDSHIP_INDEX = (select MAX(HARDSHIP_INDEX) from census_data)
```

```python
#Use a sub-query to determine the Community Area Name with most number of crimes?
%sql sqlite:///crime_data.db
%sql select COMMUNITY_AREA_NUMBER, COUNT(*) AS crime_rate \
    from crime_data \
    where COMMUNITY_AREA_NUMBER = (select MAX(COMMUNITY_AREA_NUMBER) from crime_data) \
    order by COMMUNITY_AREA_NUMBER desc
```
